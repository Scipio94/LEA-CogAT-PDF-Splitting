# CogAT PDF Splitting

Creating a script to split PDFs and assingn the local id as the file name for the ouput. The following packages were used below:
- [tabula-py](https://pypi.org/project/tabula-py/)
- [pypdf](https://pypi.org/project/pypdf/)

```python
import tabula
from tabula import read_pdf
import os
import jpype
import subprocess
import pandas as pd
import pypdf
from pypdf import PdfReader,PdfWriter
```

```python
# creating variables for area kwarg for extracting text from PDF using tabula-py
y1 = 22.27
x1 = 655.07
y2 = y1 + 7.10
x2 = x1 + 24.43
```

```python
# extracting other id from PDF using tabula-py
Other_ID = tabula.read_pdf(r"C:\Users\togarro\Downloads\Student Profile Narrative (Scoring Service Print Style).pdf",pages = 'all', area = (y1,x1,y2,x2),pandas_options = {'header': None})
```

```python
# creating dataframe from PDF
other_id_df = pd.DataFrame(pd.concat(Other_ID)) #--> concat list and then create df
other_id_df.rename(columns = {0:'Other ID'}, inplace = True) #--> renaming columns
other_id_df = other_id_df.reset_index(drop = True) #--> resetting index and dropping old index
rows = other_id_df.shape[0] #--> creating variable for the number of rows in the dataframe
```

```python
# displaying the first 5 rows of the dataframe to check indexing
other_id_df.head()
```

```python
# importing pdf 
reader = PdfReader(r"C:\Users\togarro\Downloads\Student Profile Narrative (Scoring Service Print Style).pdf")
writer = PdfWriter() #--> instantiating PDF Writer
pages = len(reader.pages) #--> returning number of pages
```

```python
# QA Statement
print(f'There are {pages} pages in the PDF file and {rows},both values are equal,{pages == rows}')
```

```python
# for loop to index, name, and split PDFs
for page_num in range(len(reader.pages)): #--> range of page numbers
    writer = PdfWriter() #--> resets the writer object in each iteration of the loop
    page_name = other_id_df['Other ID'][page_num] #--> returns other id based on indexing
    writer.add_page(reader.pages[page_num]) #--> adding pages to writer based on page number
    with open(f'{page_name}.pdf','wb') as f: #--> file writing
        writer.write(f)
        
    
   
```
